```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.feature_extraction.text import TfidfTransformer

sample_text = ["The quick brown fox jumps over the lazy dog."]
```

```python

# Initialize the CountVectorizer to get the term frequencies
count_vectorizer = CountVectorizer()
term_counts = count_vectorizer.fit_transform(sample_text)

# Get the feature names (terms)
terms = count_vectorizer.get_feature_names_out()
# Get the TF values
tf_values = term_counts.toarray() / term_counts.sum(axis=1)
print("Term Frequency (TF):")
for term, tf in zip(terms, tf_values[0]):
    print(f"{term}: {tf}")
```

```python

# Initialize the TfidfTransformer to get the IDF values
tfidf_transformer = TfidfTransformer()
idf_values = tfidf_transformer.fit(term_counts)
# print(idf_values)
# Print TF values

# Print IDF values
print("\nInverse Document Frequency (IDF):")
for term, idf in zip(terms, idf_values.idf_):
    print(f"{term}: {idf}")
```
