```python
import pandas as pd 
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("titanic.csv")
```

```python
df.columns
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
Q1=df['Age'].quantile(0.25)
Q3=df['Age'].quantile(0.75)
IQR=Q3-Q1
print("IQR(", IQR, ") =", "Q3(", Q3, ")- Q1(", Q1, ")")
```

```python
lower_limit=Q1-IQR
upper_limit=Q3+IQR
lower_limit,upper_limit
```

```python
df_without_outliers=df[(df['Age']>lower_limit)&(df['Age']<upper_limit)]
df_without_outliers
```

```python
sns.boxplot(y='Fare' , x='Sex', hue='Survived' ,data = df)
```

```python
sns.boxplot(x='Sex' , y='Fare', hue='Survived' ,data = df_without_outliers)
```

```python
sns.boxplot(x='Age' , y='Sex', hue='Survived' ,data = df)
```

```python
sns.boxplot(x='Age' , y='Sex', hue='Survived' ,data = df_without_outliers)
```
